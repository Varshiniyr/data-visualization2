# Develop a code to demonstrate Kernel Density Estimation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Marine_Fish_Data.csv')
print(df)
```

```python
fp='Fish_Population'
df=df.dropna(subset=[fp])
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(df[fp],fill=True,color='blue',bw_adjust=0.5)
plt.title(f'Kernel Density Estimation for {fp}',fontsize=14)
plt.xlabel(fp,fontsize=12)
plt.ylabel('Density',fontsize=12)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(3,3))
species_column='Species_Name'
for species in df[species_column].unique():
    sns.kdeplot(df[df[species_column]== species][fp],color='green',label=species,fill=True,alpha=0.5)
    plt.title(f'KDE for {fp} by Species',fontsize=14)
    plt.xlabel(fp,fontsize=12)
    plt.ylabel('Density',fontsize=12)
    plt.legend(title=species_column)
    plt.grid(True)
    plt.show()
```
