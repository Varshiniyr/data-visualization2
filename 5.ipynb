```python
import os
import requests
from bs4 import BeautifulSoup
```

```python
 # Function to download an image from a URL
def download_image(image_url, save_path="images"):
    # Create directory if it doesn't exist
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    # Get the image content
    response = requests.get(image_url)
    if response.status_code == 200:
        # Extract image name from URL
        image_name = image_url.split("/")[-1]
        full_path = os.path.join(save_path, image_name)
        
        # Save the image
        with open(full_path, "wb") as f:
            f.write(response.content)
        print(f"Image downloaded: {full_path}")
    else:
        print(f"Failed to download image: {image_url}")
        
# Function to parse an HTML page
def parse_html(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Extract the title
        title = soup.title.string if soup.title else "No title found"
        print(f"Page Title: {title}")
        
        # Extract all links
        print("\nLinks on the page:")
        for link in soup.find_all("a", href=True):
            print(link["href"])
    else:
        print(f"Failed to retrieve HTML from {url}")

# Function to scrape and download all images from a web page
def scrape_images(url, save_path="images"):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        image_tags = soup.find_all("img")
        
        print(f"\nFound {len(image_tags)} images. Downloading...")
        for img_tag in image_tags:
            img_url = img_tag.get("src")
            if img_url:
                # Handle relative URLs
                if not img_url.startswith("http"):
                    img_url = requests.compat.urljoin(url, img_url)
                download_image(img_url, save_path)
    else:
        print(f"Failed to scrape images from {url}")

# Main program
if __name__ == "__main__":
    # Define URLs
    html_url = "https://www.w3schools.com/html/html_images.asp"  # Replace with a target website
    image_url = "https://via.placeholder.com/150"  # Replace with an image URL
    
    # 1. Download a single image
    print("=== Downloading an Image ===")
    download_image(image_url)
    
    # 2. Parse HTML to extract title and links
    print("\n=== Parsing HTML ===")
    parse_html(html_url)
    
    # 3. Scrape and download all images from a web page
    print("\n=== Scraping Images ===")
    scrape_images(html_url)
```
