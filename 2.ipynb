## Develop a code to plot bivariate distribution considering a suitable dataset available on the open source


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
print(df.head())
```

```python
if 'Sleep Duration' in df.columns and 'Stress Level' in df.columns:
    plt.figure(figsize=(8,6))
    sns.kdeplot(data=df,x="Sleep Duration",y="Stress Level",fill=True,cmap="viridis",alpha=0.7)
    plt.title("Bivariate Distribution of Sleep Duration and Stress Level",fontsize=14)
    plt.xlabel("Sleep Duration",fontsize=12)
    plt.ylabel("Stress Level",fontsize=12)
    plt.show()
```
